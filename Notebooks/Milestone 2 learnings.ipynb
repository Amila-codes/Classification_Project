## 📊 Classification Project Phases

**Phase 1: Problem Framing (~20% of effort)**

- Define the business objective and clear success criteria
- Identify the target variable and relevant prediction timeline
- Assess data availability, label quality, and feasibility
- Consider ethical implications and fairness requirements

**Phase 2: Data Preparation (~40% of effort)**

- Collect representative, labeled data
- Analyse class distributions and feature relationships
- Address missing values, outliers, and class imbalance
- Perform train/test splits using stratified sampling where needed

**Phase 3: Model Development (~30% of effort)**

- Engineer features that reflect your domain knowledge
- Try a range of algorithms—from simple baselines to complex models
- Tune hyperparameters using cross-validation
- Evaluate with the right metrics for your problem type

**Phase 4: Deployment & Monitoring (~10% of effort)**

- Test the final model on unseen data
- Deploy the model with monitoring systems in place
- Plan for updates, retraining, and feedback loops
- Monitor for performance drift, bias, and ethical compliance

**🚀 Success Factors**

- Involve domain experts early in the problem-framing process
- Start simple—baseline models help set expectations
- Plan for long-term maintenance from the beginning
- Revisit ethical considerations at every stage, not just at the end

## Risk Score Formula
R=S×LR = S \times L<br>
R=S×L<br>
Where:<br>
R (Risk Score) → overall level of risk<br>
S (Severity) → how serious the impact would be (1 = minor, 5 = catastrophic)<br>
L (Likelihood) → how likely it is to happen (1 = rare, 5 = very likely)<br>

## Success Metrics Aligned with Risk

Model selection should minimize expected real-world cost, not simply optimize a convenient metric.<br>
For binary classification with:<br>
- C FP  = cost of a false positive
- C FN = cost of a false negative

The expected monetary cost (EMC) is defined as:<br>
EMC = P_FP * C_FP + P_FN * C_FN<br>

Where:<br>
- P_FP = False Positive Rate (FPR)  
- P_FN = False Negative Rate (FNR)  

These values come from your confusion matrix.<br>
Your primary metric should align with the most costly error type.<br>

Error-Cost Regime Guide

Error-Cost Regime                | Primary Metric        | Why?
--------------------------------|----------------------|------------------------------------------
C_FN >> C_FP                    | Recall (TPR)         | Missing true cases is catastrophic
C_FP >> C_FN                    | Precision (PPV)      | False alarms waste resources
Balanced                        | F1 or Matthews Corr. | Balances both error types
Ranking focus                   | ROC-AUC / PRC-AUC    | Rank-order by confidence/risk

Note: The optimal threshold t∗  is rarely 0.5.

**🌍 Real-World Impact Example**<br>

In 2024, the U.S. Wildfire Early-Alert System set a hard constraint:<br>
Recall ≥ 0.92<br>
Why? An EMC analysis revealed:<br>
- Missing a fire (FN) costs 2 million usd<br>
- Raising a false alarm (FP) costs ~$15,000<br>

Prioritising recall significantly reduced catastrophic loss with minimal operational overhead.<br>

## Time Horizon & Prediction Frequency-Guard Against Leakage

Every classification project runs on a timeline. You make a prediction at time t0 , for something that may happen by t0+h (the prediction horizon).<br> All input features must be collected before the prediction time.
-  Using future data (after t0) causes label leakage<br>
-  Using overly old data adds noise<br>

**🌍 Real-World Case Study**<br>
The UK National Grid’s D-2 Failure Forecaster makes predictions every 30 minutes, with a 48-hour horizon to pre-empt substation failures.<br>
If the horizon is too short, repairs can't be scheduled<br>
If it’s too long, predictions become noisy<br>
Their optimal design of h=48h = 48h=48 hours led to a 22% reduction in outage minutes (UK National Grid, 2024).<br>

## 3 classes for the classification project

1. High Risk (Likely to Churn)

Action:
Customers in this category will receive targeted retention interventions such as discounts, personalized offers, or direct outreach from customer support teams.

Rationale:
These customers have a high probability of churning within the prediction horizon. Proactively engaging them is critical to prevent immediate revenue loss and improve retention rates.

2. Medium Risk (At Risk / Uncertain)

Action:
Customers will be monitored and engaged through low-cost strategies such as reminder emails, product recommendations, or usage nudges.

Rationale:
These customers show some indicators of churn risk but are not yet critical. Early, lightweight engagement can prevent them from transitioning into high-risk customers while minimizing operational costs.

3. Low Risk (Stable Customers)

Action:
No immediate intervention is required. Customers will continue receiving the standard service experience.

Rationale:
These customers are unlikely to churn in the near future. Avoiding unnecessary incentives helps optimize resource allocation and maintain profitability.

    Predict whether an individual customer will churn (i.e., cancel their subscription or become inactive) within the next 30 days, based on data available at the current time. The prediction informs whether to offer a retention incentive such as a discount, personalized outreach, or service improvement. Each prediction is made at the customer level, with churn as the outcome event. By identifying at-risk customers within this time horizon, the business can take timely action to improve retention and reduce revenue loss.
      
2. 📈 Success Metrics (~50 words + table)
    - Set baseline and aspirational targets
    - Use stakeholder interviews or industry benchmarks
    - Use placeholders now; replace with real numbers in Milestone 3
      
3. 👥 Stakeholders & Impact (~100 words + quadrant visual)
    - Map direct/indirect beneficiaries and harms
    - Qualitative only—no data needed
      
4. 🗃️ Data Overview (~120 words)
    - List anticipated sources (e.g., "TrashNet images from Kaggle")
    - Estimate the 5 V’s: Volume, Variety, Velocity, Veracity, Value
    - Flag uncertainties for validation later
      
5. 🧠 Model Scope (~60 words)
    - Propose one baseline model (e.g., logistic regression)
    - Add a stretch model for higher performance (e.g., ensemble, fine-tuned CNN)
      
6. ⚖️ Ethics & Compliance (~80 words)
    - Identify sensitive attributes (if any)
    - Estimate your AI risk tier (e.g., low under EU-AI Act)
    - No data exploration needed—just domain reasoning
      
7. 🚀 Deployment Context (~60 words)
    - Define runtime setting: real-time or batch
    - Note latency budget and whether predictions will be human-reviewed
    - All based on intended use, not dataset
