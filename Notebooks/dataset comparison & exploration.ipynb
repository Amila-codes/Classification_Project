# Data sets Pre analysis

## Functions

```python
# Define sensitive_data function
def sensitive_data(df):
  """Check sensitive data columns"""
  sus_kw = ["gender","race","ethnicity","pregnancy","zipcode","religion"]
  sensitive_cols = [c for c in df.columns if any(k in c.lower() for k in sus_kw)]
  return sensitive_cols
```

## Telcom Churn

```python
# Load telcom data set
df_telcom = pd.read_csv(r"..\Data\Raw\Telco_Cusomer_Churn.csv")
display(df_telcom.head())
df_telcom.shape
```

```python
# Check target proportions
print(df_telcom['Churn'].value_counts(normalize=True))
```

```python
df_telcom.columns
```

```python
# Call the function
result = sensitive_data(df_telcom)
print(f"Potentially sensitive: {result}")
```
