## Problem Statement & Context

Customer churn is a major challenge for companies that rely on subscription-based services such as Internet Service Providers (ISPs). When customers leave, companies not only lose recurring revenue but also incur additional costs to acquire new users.

This project focuses on a binary classification task:

Predict whether a customer is likely to churn (True) or remain (False).

Customer churn affects multiple stakeholders. Customers may experience declining service quality or unmet expectations before leaving, while customer success and marketing teams need to identify at-risk users early enough to intervene. For businesses, high churn rates directly impact profitability and long-term growth.

A reliable churn prediction model can support better decision-making by enabling teams to proactively target customers who are likely to leave. This includes offering personalized incentives, improving support interactions, and optimizing retention strategies. By focusing efforts on the most at-risk customers, companies can allocate resources more efficiently and improve overall customer experience.

## Hypothesis & Intended Impact

This project is based on the assumption that customer behavior, usage patterns, and account-related information contain enough signal to predict churn accurately. By analyzing these features, the model should be able to identify customers who are at risk of leaving before the actual event occurs.

If the model performs well, it can enable more proactive and targeted retention strategies. Instead of applying broad and often inefficient campaigns, companies can focus on customers who are most likely to churn. This can lead to improved customer satisfaction, as users receive more relevant support and offers tailored to their needs.

From a business perspective, better churn prediction can reduce customer acquisition costs, increase customer lifetime value, and improve overall operational efficiency. More broadly, focusing on retaining existing customers rather than constantly replacing them supports more sustainable and resource-efficient business practices.

## Dataset Discovery & Selection

To support this classification task, three datasets were initially considered: a telecom dataset, a bank customer churn dataset, and an ISP dataset. Each dataset was evaluated based on alignment with the problem, data quality, scale, documentation, and ethical considerations.

The ISP dataset was ultimately selected because it provides a strong balance across these dimensions. It is well-documented, sufficiently large, and directly aligned with the problem of predicting customer churn in a subscription-based service context. In addition, the dataset is released under a public domain (CC0) license, which removes potential legal or usage constraints.

Compared to the alternatives, the ISP dataset offers high-quality features, reliable labels, and minimal access barriers. While the bank dataset also performed well in evaluation, the ISP dataset was chosen due to its closer relevance to the specific churn scenario being modeled.

## Dataset Justification & Target Variable

The ISP dataset is well-suited for this problem because it includes detailed information about customer behavior, service usage, and account history. These types of features are commonly associated with churn and provide a strong foundation for building predictive models.

The target variable in this dataset is Churn, which indicates whether a customer has left the service. It is defined as a binary variable:

True: the customer has churned
False: the customer has remained

This target is directly relevant to the business problem and is both meaningful and actionable. Predicting churn allows stakeholders to intervene before the customer leaves, making it possible to influence outcomes rather than simply observe them.

## Features Overview

The dataset contains a mix of numerical, categorical, and behavioral features that describe different aspects of customer activity.

Numerical features include variables such as usage levels and billing amounts, which can indicate how actively a customer is using the service. Categorical features describe subscription types or service configurations, which may influence customer satisfaction. Behavioral features capture patterns over time, such as changes in usage or interaction frequency. Account-related features, such as tenure, provide context about the customer’s relationship with the service.

Several features are expected to be particularly informative. For example, customers with shorter tenure may be more likely to churn, while unusual changes in usage or billing behavior may signal dissatisfaction. Overall, the feature set appears rich enough to support accurate predictions.

## Data Overview

The dataset contains approximately 72,000 customer records, making it large enough to support robust model training and evaluation.

The target variable is reasonably well distributed:

- Around 40,050 customers have churned
- Around 32,224 customers have not churned

This indicates a moderate class balance, which is beneficial for classification tasks. Since neither class is extremely underrepresented, the model can learn patterns for both churners and non-churners without requiring heavy rebalancing techniques.

## Ethical Considerations

Ethical considerations were taken into account during the data preparation stage. One feature, gender, was identified as a sensitive attribute and removed from the dataset. This decision was made to reduce the risk of introducing bias into the model and to avoid predictions being directly influenced by protected characteristics.

Even after removing explicit sensitive features, there is still a possibility of indirect bias. Some variables may act as proxies for demographic information, which could lead to unfair outcomes for certain groups. This highlights the importance of evaluating fairness in later stages of the project.

From a regulatory perspective, this system can be considered a limited-risk application, as it supports decision-making but does not directly automate high-stakes outcomes. However, careful monitoring is still necessary to ensure responsible use.

## Real-World Framing

The prediction task can be clearly defined in a real-world context.

The unit of analysis is an individual customer, and the event of interest is whether that customer churns. Predictions are made with a horizon of 30 days, meaning the model estimates the likelihood of churn within the next month. These predictions can be generated on a daily basis using the most recent available data.

The outputs of the model are intended to support actions such as targeted retention campaigns, personalized offers, or proactive customer support. The main stakeholders include customer success teams, marketing teams, and the customers themselves.

There is an important trade-off between recall and precision. In this context, recall is especially important, as failing to identify a customer who is about to churn can result in lost revenue. However, improving recall may lead to more false positives, which can increase the cost of retention efforts.

## Success Metrics

The primary evaluation metric for this project is recall for the churn class. This reflects the importance of identifying as many at-risk customers as possible.

Additional metrics such as precision and F1-score will also be considered to provide a more balanced view of model performance.

Beyond model-level metrics, success can also be measured using business outcomes. These include a reduction in churn rate, improved effectiveness of retention campaigns, and an increase in customer lifetime value. Together, these metrics provide a more complete picture of the model’s real-world impact.

## Functions Used

```python
# Define detect_variable_types function
def detect_variable_types(df, target_col=None):
    """Automatically detect variable types in a DataFrame"""
    types = {}
    
    for col in df.columns:
        if col == target_col:
            types[col] = 'target'
        elif pd.api.types.is_datetime64_any_dtype(df[col]):
            types[col] = 'datetime'
        elif pd.api.types.is_numeric_dtype(df[col]):
            if df[col].nunique() == 2:
                types[col] = 'binary'
            else:
                types[col] = 'numerical'
        elif df[col].nunique() < 20:  # Threshold for categorical*
            types[col] = 'categorical'
        elif df[col].dtype == 'object':
            # Check if it's text (long strings) or categorical*
            avg_length = df[col].astype(str).str.len().mean()
            if avg_length > 50:
                types[col] = 'text'
            else:
                types[col] = 'categorical'
    
    return types
```

```python
# Check missing values
def missing_values(df):
    missing_v = df.isnull().sum()
    missing_percentage = (missing_v / len(df)) * 100
    zero_v = (df.loc[:, df.nunique() > 2] == 0).sum()
    zero_v_percentage = (zero_v/len(df))*100

    summary = pd.DataFrame({
        "Missing values": missing_v,
        "Missing %": missing_percentage,
        "Zero values": zero_v,
        "Zero %": zero_v_percentage
    })

    print("Missing value summary:\n")

    return summary
```

```python
# Check range and distribution
# Quick sanity checks for numerical data
def basic_range_check(df):
    numeric_cols = df.select_dtypes(include=['number']).columns
    
    for col in numeric_cols:
        print(f"{col}: min={df[col].min()}, max={df[col].max()}")
        
        # Flag obviously problematic values
        if col.lower() in ['subscription_age'] and (df[col].min() < 0 or df[col].max() > 150):
            print(f"  ⚠️ Suspicious subscription_age values detected")
        elif col.lower() in ['bill_avg', 'download_avg', 'upload_avg', 'download_over_limit', 'remaining_contract'] and df[col].min() < 0:
            print(f"  ⚠️ Negative financial values detected")
    
```

```python
# Check duplicates
def duplicates(df, key=None):
  
  exact_dup = df.duplicated().sum()
  print(f"exact duplicates: {exact_dup}")

  key_dup = None
  if key is not None:
    key_dup = df.duplicated(subset=key).sum()
    print(f"key_duplicates: {key_dup}")
  
  return (exact_dup, key_dup)
```
