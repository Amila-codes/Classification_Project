## Evaluation Before Download

Check	What to Look For
1. **Alignment**	Do the classes match your framed problem?
2. **Label veracity**	Who labelled the data? Experts, crowd, or automated scripts?
3. **Scale**	Enough examples per class? (≥50 per class is a pilot minimum)
4. **Class balance**	Are labels skewed? Will you need to re-sample?
5. **File integrity**	Sample 1% through previews; check for broken or corrupted files
6. **Documentation**	Is there a README, data dictionary, and license?
7. **Ethics**	Any sensitive content? Faces, personal info, endangered species locations?
8. **Access hurdles**	IRB approval, API key, or license restrictions (e.g. non-commercial use)?

Score each dimension from 1–5.<br>
If alignment or ethics score ≤3 — skip the dataset.<br>
🧪 Mini-Example (Text Dataset)<br>
Dataset: "Global Spam SMS" (Kaggle)<br>
✅ Alignment = 5 → Binary spam vs ham.<br>
✅ Label veracity = 4 → Crowdsourced with consensus.<br>
✅ Scale = 5 → Over 5,500 messages.<br>
⚠️ Ethics = 2 → Contains phone numbers → hash before use.<br>
Decision: Download with preprocessing to protect privacy.<br>

## Licensing Demystified

How to Check<br>
Look for a LICENSE or NOTICE file<br>
Check the Usage or Terms section of the repository<br>
If no license is present, assume the dataset is not open until clarified<br>

Common License Types<br>
License	What It Means<br>
CC0 / Public Domain	✅ Safest — no restrictions<br>
ODC-BY	⚠️ Requires attribution (Open Data Commons)<br>
CC-BY-SA	⚠️ Share-alike: you must publish derivatives under the same license<br>
CC-BY-NC	🚫 No commercial use — can’t be embedded in paid tools or APIs<br>
Custom Research License	🧩 Often includes clauses against resale or external redistribution — always check the fine print<br>

🚀 Capstone Hook<br>
